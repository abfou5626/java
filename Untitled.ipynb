```python
def transliterate_french_to_arabic(text):
    mapping = {
        'a': 'ا',
        'b': 'ب',
        'c': 'س',
        'd': 'د',
        'e': 'ا',
        'f': 'ف',
        'g': 'ج',
        'h': 'ه',
        'i': 'ي',
        'j': 'ج',
        'k': 'ك',
        'l': 'ل',
        'm': 'م',
        'n': 'ن',
        'o': 'و',
        'p': 'ب',
        'q': 'ك',
        'r': 'ر',
        's': 'س',
        't': 'ت',
        'u': 'و',
        'v': 'ف',
        'w': 'و',
        'x': 'كس',
        'y': 'ي',
        'z': 'ز'
    }

    transliterated_text = ''.join(mapping.get(char.lower(), char) for char in text)
    return transliterated_text

# Exemple d'utilisation
source_text = "abdoul"
transliterated_text = transliterate_french_to_arabic(source_text)
print(f"Texte source : {source_text}")
print(f"Texte translittéré : {transliterated_text}")

```

```python
from transliterate import translit

def transliterate_name(name, source_language, target_language):
    try:
        transliterated_name = translit(name, source_language, target_language)
        return transliterated_name
    except Exception as e:
        print(f"Erreur lors de la translittération : {e}")
        return None

# Exemple d'utilisation
source_name = "Александар"
source_language = "bg"
target_language = "sr"  # Langue cible (par exemple, russe)

transliterated_name = transliterate_name(source_name, source_language, target_language)
print(f"Nom source : {source_name}")
print(f"Nom translittéré : {transliterated_name}")

```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Flatten, Dense
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
fashion_mnist = keras.datasets.fashion_mnist
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()
```

```python
X_valid, X_train = X_train_full[:5000] / 255.0, X_train_full[5000:] / 255.0
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
```

```python
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat","Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
 class_names[y_train[0]]
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=[28, 28]))
model.add(keras.layers.Dense(300, activation="relu"))
model.add(keras.layers.Dense(100, activation="relu"))
model.add(keras.layers.Dense(10, activation="softmax"))
```

```python
 model.layers
```

```python
hidden1 = model.layers[1]
weights, biases = hidden1.get_weights()
```

```python
weights.shape
```

```python
biases.shape
```

```python
model.compile(loss="sparse_categorical_crossentropy",optimizer="sgd",metrics=["accuracy"])
```

```python
 history = model.fit(X_train, y_train, epochs=30,validation_data=(X_valid, y_valid))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1) # Régler la plage verticale sur [0-1]
plt.savefig('diagramme.png')
plt.show()
```
